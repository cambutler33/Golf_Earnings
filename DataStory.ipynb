```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
df = pd.read_csv("GolfEarnings.csv")
df= df[df['Putts Per Hole'] != 0.0]
df= df[df['Cuts Made'] != 0.0]
df = df.drop(columns=(['Rounds Played', 'Cuts Made', 'Top 10 Finishes']))
```

```python
dfCorr = df.corr()
dfCorr = dfCorr.drop(dfCorr.index[1:])
dfCorr
```

```python
dfCorr = dfCorr.abs()
dfCorr
```

```python
#sorted_df = df.sort_values(df.last_valid_index(), axis=1)
sort_dfC = dfCorr.sort_values(dfCorr.last_valid_index(), axis=1, ascending=False)
sort_dfC
```

```python
df1 = sort_dfC.drop(columns=(['Earnings for Year', 'Wins']))
df1
```

```python
sns.barplot(data=df1, color='forestgreen')
```

```python
sns.set_theme(style='darkgrid', rc={'figure.dpi': 147},              
              font_scale=0.7)
fig, ax = plt.subplots(figsize=(7, 2))
ax.set_title('Correlation to Earnings Per Year on PGA Tour')
ax.set_ylabel('Absolute Value of Correlation')
sns.barplot(data=df1, ax=ax, color='forestgreen');
```

```python
import textwrap
def wrap_labels(ax, width, break_long_words=False):
    labels = []
    for label in ax.get_xticklabels():
        text = label.get_text()
        labels.append(textwrap.fill(text, width=width,
                      break_long_words=break_long_words))
    ax.set_xticklabels(labels, rotation=0)
```

```python
wrap_labels(ax, 10)
ax.figure
```
