```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
df = pd.read_csv("GolfEarnings.csv")
df= df[df['Putts Per Hole'] != 0.0]
df= df[df['Cuts Made'] != 0.0]
df = df.drop(columns=(['Rounds Played', 'Cuts Made', 'Top 10 Finishes']))
```
